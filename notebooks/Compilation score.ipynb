```python
from joblib import dump,load
df_score_DL = load("df_score_DL.joblib")
```

```python
df_score_DL_franck = load("df_score_DL_franck.joblib")
```

```python
df_score_DL_fred_2 = load("df_score_DL_fred_2.joblib")
df_score_DL_fred_2
```

```python
import pandas as pd
df = pd.concat([df_score_DL,df_score_DL_franck,df_score_DL_fred_2],axis=0, ignore_index=True)
```

```python
df= df.drop(columns=['optimizer','fonction loss','model','DataGenerator'])
```

```python
df['Duree_entrainement_heure'] = df['Duree_entrainement en sec'].apply(lambda x:f"{int(x//3600)}h{int((x - 3600 * (x//3600)) // 60)}min")
```

```python
df= df.drop(columns=['Duree_entrainement en sec'])
```

```python
df["test_acc"]=df["test_acc"].fillna("Non calculé")
df = df.loc[df["test_acc"]!="Non calculé" ]
df["acc_test"] = df["test_acc"].apply(lambda x:x[1])
```

```python
df = df.sort_values(by='acc_test', ascending=False)
```
