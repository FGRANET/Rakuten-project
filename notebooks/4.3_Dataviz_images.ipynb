```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
```

# Fonctions

```python
# fonction pour récupérer le chemin absolu d'un dossier placé au même endroit que le notebook en cours
def get_output_path_folder(output_foldername):
    notebook_filename = os.path.abspath("__file__")
    notebook_directory = os.path.dirname(notebook_filename)
    return os.path.join(notebook_directory, output_foldername)

def selectRandom(liste):
    return random.choice(liste)
```

## Afficher une image du jeu d'entrainement

```python
#chemin où sont stockées les images d'entrainement classées 
repertoire_train = get_output_path_folder("images_traitees_classe_train")

# Chargement de l'image et affichage
path = repertoire_train + "\\1160\\image_114783730_product_3530750.jpg"

img_color = cv2.imread(path, cv2.IMREAD_COLOR)
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB) #passer du système de couleur BGR à RBG

fig = plt.figure(figsize=(8,5))
ax = plt.subplot(121)
plt.imshow(img_color) # affichage de l'image

# Suppression des axes
ax.set_axis_off()
    
# Titre de la figure
plt.title('Une photo initiale avant transformation')
plt.show();  
```
